```python
#Sistema de Calificaciones CORE

asignaturas_permitidas = [
    "matematicas", "fisica", "quimica", "historia",
    "lenguaje", "educacion fisica", "ingles", "ciencias"]

"""
primero se debe ingresar el numero de estudiantes
"""

def obtener_numero_estudiantes():
    while True:
        try:
            num_estudiantes= int(input("ingrese el numero de estudiantes: "))
            if num_estudiantes <= 0:
                print("El número de estudiantes ingresado debe ser mayor a 0")
                continue
            return num_estudiantes
        except ValueError:
           print("Por favor, ingrese un número válido.")


"""
Función para obtener el nombre del estudiante
"""

def obtener_nombre_estudiante():
    while True:
        nombre= input("ingrese el nombre del estudiante: ").strip() #Utilizo strip para eliminar espacios vacios
        if nombre:
            return nombre
        else:
            print("El nombre del estudiante no puede quedar en blanco.")
            continue

"""
funcion para obtener el numero de asignaturas """

def obtener_numero_asignaturas():
  while True:
    try:
        num_asignaturas=int(input("Ingrese el numero de asignaturas: "))
        if num_asignaturas <= 0:
           print("El número ingresado debe ser mayor a 0")
           continue
        return num_asignaturas
    except ValueError:
        print("Por favor, ingrese un número válido.")

"""
funcion para obtener las calificaciones """

def obtener_calificaciones(num_asignaturas):
    calificaciones=[]
    for i in range(num_asignaturas):
        while True:
            asignatura= input(f"ingrese el nombre de la asignatura {i+1}: ").strip().capitalize()
            if not asignatura:
                print("El nombre de la asignatura no puede quedar sin datos. Intente nuevamente.")

            if asignatura.lower() not in asignaturas_permitidas:
                print("Asignatura no permitida. Intente nuevamente.")
                continue
            elif asignatura.isdigit():
                print("El nombre de la asignatura no puede ser un número. Intente nuevamente.")
            else:
                break #Sale del while si es correcto

        while True:
            try:
                calificacion= float(input(f"ingrese la calificacion de {asignatura}: "))
                if 1 <= calificacion <= 10:
                    calificaciones.append((asignatura, calificacion))
                    break
                else:
                    print("La calificación debe ser entre 1 y 10. Intente nuevamente")
            except ValueError: #Esto siempre alineado con try
                print("Error: Por favor, ingrese un número válido.")
    return calificaciones

"""
funcion para calcular el promedio de
las calificaciones ingresadas """

def calcular_promedio(calificaciones):

  total_calificaciones = sum(calificacion for _, calificacion in calificaciones)
  promedio= total_calificaciones / len(calificaciones)
  return promedio

"""
funcion Aprueba o Reprueba basándonos en el promedio de sus calificaciones """

def determinar_estado(promedio):

  if promedio >= 6.0:
    return "Aprobado"
  else:
    return "Reprobado"

"""
 Función para generar resumen por estudiante"""

def imprimir_resumen(estudiantes):

   print("\nResumen de Estudiantes:")
   for estudiante in estudiantes:
    print(f"Nombre: {estudiante['nombre']}")
    print(f"Promedio: {estudiante['promedio']:.1f}") #.1f indica la cantidad de decimales que imprime
    print(f"Estado: {estudiante['estado']}")
    print()

#Función principal para procesar la información:

num_estudiantes = obtener_numero_estudiantes()
estudiantes = []

for _ in range(num_estudiantes):
    nombre = obtener_nombre_estudiante()
    num_asignaturas = obtener_numero_asignaturas()
    calificaciones = obtener_calificaciones(num_asignaturas)
    promedio = calcular_promedio(calificaciones)
    estado = determinar_estado(promedio)

    estudiantes.append({
        'nombre': nombre,
        'promedio': promedio,
        'estado': estado
    })

imprimir_resumen(estudiantes)
```
